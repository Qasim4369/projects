```python
## Data Cleaning
df.isnull().any()
```

```python
df.duplicated().any()
## Duplicates also exists so its better to drop them
```

```python
df=df.drop_duplicates()
```

```python
## Checking if there is any row with only null values
df.isnull().all(axis=1).any()
```

```python
df=df.dropna()
```
