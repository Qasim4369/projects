## Image Thresholding

```python
img = cv2.imread('download.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
##Types of Thresholdings
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)
  
# the window showing output images
# with the corresponding thresholding

cv2.imshow('Binary Threshold', thresh1)
cv2.imshow('Binary Threshold Inverted', thresh2)
cv2.imshow('Truncated Threshold', thresh3)
cv2.imshow('Set to 0', thresh4)
cv2.imshow('Set to 0 Inverted', thresh5)
    
# De-allocate any associated memory usage  
if cv2.waitKey(0) & 0xff == 27: 
    cv2.destroyAllWindows()
```

## Morphological Operations

```python
## To perform Morphological Operations Binary Images are used
img = cv2.imread('img.jpg',0)
binr = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
```

```python
## Define Kernel
kernel = np.ones((5,5),np.uint8)
## Invert Image
invert = cv2.bitwise_not(binr)
## Erode Image
erosion = cv2.erode(invert,kernel,iterations = 1)
```

```python
#print inverted output
plt.imshow(invert,cmap='gray')
```

```python
## Eroded image is thinner than the original one
plt.imshow(erosion,cmap='gray')
```

```python
## Dilation Operation
dilation = cv2.dilate(invert , kernel ,iterations=1)
## Dilation operation is thicker than the original image
plt.imshow(dilation,cmap='gray')
```

```python
## Morphological Gradient
## Applies and computes the difference between eroded and dilated image
morph_gradient = cv2.morphologyEx(invert,cv2.MORPH_GRADIENT,kernel)
plt.imshow(morph_gradient,cmap='gray')
```
